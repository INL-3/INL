```python
import os
import json


with open("data.json", "r", encoding="utf-8") as file:
    data = json.load(file)

# konfiguracja klienta(Klucz API)
os.environ["GEMINI_API_KEY"] = data["geminiApiKey"]
```

```python
documents = [
    "Mieszko I był księciem Polan i zapoczątkował państwo polskie.",
    "Bolesław Chrobry został pierwszym królem Polski w 1025 roku.",
    "Koronacja Bolesława Chrobrego odbyła się w Gnieźnie."
]

question = "Kto był pierwszym królem Polski?"
```

```python
import google.generativeai as genai

# Wybór modelu
model = genai.GenerativeModel("gemini-2.5-flash")

def gemini_answer(question: str, documents: list[str]) -> str:

    context = "\n".join(documents)

    prompt = f"""
Jesteś systemem odpowiadającym na pytania na podstawie dostarczonego kontekstu.

Odpowiadaj WYŁĄCZNIE na podstawie dostarczonego kontekstu.
Jeśli odpowiedź NIE znajduje się w kontekście, odpowiedz dokładnie:
"Nie wiem".

Kontekst:
\"\"\"
{context}
\"\"\"

Pytanie:
{question}

Odpowiedź:
- Możliwie jak najkrótsza odpowiedź
- bez dodatkowych komentarzy
"""

    response = model.generate_content(prompt)

    return response.text.strip()
```

```python
answer = gemini_answer(question, documents)

print("PYTANIE:")
print(question)

print("\nODPOWIEDŹ Gemini:")
print(answer)
```
